```python
def approx_factorial(n):
    return sqrt(2.*pi*n)*((n/e)**n)*(1. + (1./(12.*n)) + (1./(288.*n*n)) - (139./(51840.*n*n*n)) - (571./(2488320.*n*n*n*n)))

def fnspfact(x):
    return 1./((x+2.)*(x+1.)*x*(x-1.))

def afact(n,m):
    if n==0:
        logn=0.
    else:
        logn=n*log(n)-n + 0.5*log(2.*pi*n) + log(1. +  (1./(12.*n)) + (1./(288.*n*n)) - (139./(51840.*n*n*n)) - (571./(2488320.*n*n*n*n)))
    
    if m==0:
        logm=0.
    else:
        logm=m*log(m)-m + 0.5*log(2.*pi*m) + log(1. +  (1./(12.*m)) + (1./(288.*m*m)) - (139./(51840.*m*m*m)) - (571./(2488320.*m*m*m*m)))
    print logn,logm
    return exp(logn - logm)

def calc_factorial_log(x):
    '''Uses the Stirling approximation to calculate the natural log of the factorial'''
    if x==0.:
        logxf=0.
    else:
        logxf=x*log(x)-x + 0.5*log(2.*pi*x) + log(1. +  (1./(12.*x)) + (1./(288.*x*x)) - (139./(51840.*x*x*x)) - (571./(2488320.*x*x*x*x)))
    return logxf

def return_nlm(l,m):
    '''Returns sqrt((l-m)!/(l+m)!)'''
    y=0.5*(calc_factorial_log(l-m) - calc_factorial_log(l+m) + calc_factorial_log(l-2) - calc_factorial_log(l+2))
    return exp(y)
```
