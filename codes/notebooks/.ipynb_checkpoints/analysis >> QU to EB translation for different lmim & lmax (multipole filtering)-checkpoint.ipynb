```python
respath="../../figures/simulated/multipole_filter/"
figpath=respath
!mkdir $figpath

datapath = respath + "generated_data/"
!mkdir $datapath

genplots=True
```

```python
nside=64 ; upnside=nside ; npix=h.nside2npix(upnside)
lmax=3*nside ; lmin=2
ell=arange(lmax+1)
fll=ell*(ell+1)/(2.*pi)

odisc=min(180.,180.*(32./lmax))
print odisc
```

```python
from matplotlib import rcParams,rc
params = {'backend': 'pdf',
          'savefig.dpi': 200,
          'axes.labelsize': 15,
          'text.fontsize': 15,
          'axes.linewidth' : 2,
          'lines.linewidth' : 2,
          'font.size': 15,
          'xtick.labelsize': 15,
          'ytick.major.pad': 6,
          'xtick.major.pad': 6,
          'ytick.labelsize': 15,
          'text.usetex': True,
          'font.family':'sans-serif',
          'font.sans-serif':'FreeSans'}
rc('text.latex', preamble='\usepackage{sfmath}')
rcParams.update(params)
```

```python
cltemp=h.read_cl("../../data/planck_wp_highL_lensing_param.fits")
cl=[cltemp[0],cltemp[1],cltemp[2],cltemp[3]*0] # Reverse E and B spectra

#cl=[cltemp[0],cltemp[2],cltemp[1],cltemp[3]*0] # Reverse E and B spectra

random.seed(0) ; tqu=h.synfast(cl,nside,pol=True,new=True,verbose=False)

# Get teb maps from Healpix.
alm=h.map2alm(tqu,lmax=3*nside,pol=True) ; teb=h.alm2map(alm,upnside,pol=False,verbose=False)

# Get data power spectrum.
cld=h.alm2cl(alm)

# Get Q/U corresponding to E and B modes.
alm[2][:]=0.0 ; equ=h.alm2map(alm,nside=nside,pol=True,verbose=False)
alm=h.map2alm(tqu,lmax=3*nside,pol=True) ; alm[1][:]=0.0 ; bqu=h.alm2map(alm,nside=nside,pol=True,verbose=False)

fl=ones(3*nside+1,float) ; fl[:lmin+1]=0
```

```python
y1=rk.real_space_queb_kernels(nside,tmax=2.2*odisc,maxlmax=3*nside,sampling=2500)
y2=rk.real_space_queb_kernels(nside,tmax=2.2*odisc,maxlmax=3*nside,sampling=2500)
y3=rk.real_space_queb_kernels(nside,tmax=2.2*odisc,maxlmax=3*nside,sampling=2500)
```

```python
lmin=2;lmax=3*nside
y1.calc_qu2eb_rad_kernel(lmax,lmin) ; y1.calc_qu2queb_rad_kernel(lmax,lmin)

lmin=64;lmax=3*nside#-32
y2.calc_qu2eb_rad_kernel(lmax,lmin) ; y2.calc_qu2queb_rad_kernel(lmax,lmin)

lmin=128;lmax=3*nside#-64
y3.calc_qu2eb_rad_kernel(lmax,lmin) ; y3.calc_qu2queb_rad_kernel(lmax,lmin)
```

```python
figure()
plot(y3.theta*180/pi,(y1.theta**1.)*y3.rad_ker/max(y3.rad_ker),"b-",lw=2,label=r"$\ell_{\rm min}=128~;~\ell_{\rm max}=192$",alpha=0.7)
plot(y2.theta*180/pi,(y1.theta**1.)*y2.rad_ker/max(y2.rad_ker),"g-",lw=1.5,label=r"$\ell_{\rm min}=64~;~\ell_{\rm max}=192$",alpha=0.8)
plot(y1.theta*180/pi,(y1.theta**1.)*y1.rad_ker/max(y1.rad_ker),"r-",lw=0.7,label=r"$\ell_{\rm min}=2~;~\ell_{\rm max}=192$")
hlines(0,1e-3,60)
xlim(1e-2,60)
legend(loc=0)
legend(loc=0,fontsize=10)
xlabel(r"$\beta$ degrees")
ylabel(r"$\beta*{\rm f}_{\ell}(\beta)$")
#semilogx()
savefig(figpath + "f_rad_ker_fn_vary_lmin.pdf",dpi=300,bbox_inches="tight")


figure()
plot(y3.theta*180/pi,(y1.theta**1.)*y3.rad_ker_m2/max(y3.rad_ker_m2),"b-",lw=2,label=r"$\ell_{\rm min}=128~;~\ell_{\rm max}=192$",alpha=0.7)
plot(y2.theta*180/pi,(y1.theta**1.)*y2.rad_ker_m2/max(y2.rad_ker_m2),"g-",lw=1.5,label=r"$\ell_{\rm min}=64~;~\ell_{\rm max}=192$",alpha=0.8)
plot(y1.theta*180/pi,(y1.theta**1.)*y1.rad_ker_m2/max(y1.rad_ker_m2),"r-",lw=0.7,label=r"$\ell_{\rm min}=2~;~\ell_{\rm max}=192$")
hlines(0,1e-3,60)
xlim(1e-2,60)
legend(loc=0)
legend(loc=0,fontsize=10)
xlabel(r"$\beta$ degrees")
ylabel(r"$\beta *{}_{-2}{\rm f}_{\ell}(\beta)$")
#semilogx()
savefig(figpath + "fm2_rad_ker_fn_vary_lmin.pdf",dpi=300,bbox_inches="tight")

figure()
plot(y3.theta*180/pi,(y1.theta**1.)*y3.rad_ker_p2/max(y3.rad_ker_p2),"b-",lw=2,label=r"$\ell_{\rm min}=128~;~\ell_{\rm max}=192$",alpha=0.7)
plot(y2.theta*180/pi,(y1.theta**1.)*y2.rad_ker_p2/max(y2.rad_ker_p2),"g-",lw=1.5,label=r"$\ell_{\rm min}=64~;~\ell_{\rm max}=192$",alpha=0.8)
plot(y1.theta*180/pi,(y1.theta**1.)*y1.rad_ker_p2/max(y1.rad_ker_p2),"r-",lw=0.7,label=r"$\ell_{\rm min}=2~;~\ell_{\rm max}=192$")
hlines(0,1e-3,60)
xlim(1e-2,60)
legend(loc=0)
legend(loc=0,fontsize=10)
xlabel(r"$\beta$ degrees")
ylabel(r"$\beta * {}_{+2}{\rm f}_{\ell}(\beta)$")
#semilogx()
savefig(figpath + "fp2_rad_ker_fn_vary_lmin.pdf",dpi=300,bbox_inches="tight")
```

```python
disc=2.*odisc 
y1.setup_taper(disc,3.) ; y2.setup_taper(disc,3.) ; y3.setup_taper(disc,3.)
y1.setup_fn_rad_ker() ; y2.setup_fn_rad_ker() ; y3.setup_fn_rad_ker()
```

```python
rteb0=y1.convert_qu2eb(tqu,discsize=disc)
requ0,rbqu0=y1.decompose_qu2_equ_bqu(tqu,discsize=disc)
```

```python
rteb1=y2.convert_qu2eb(tqu,discsize=disc)
requ1,rbqu1=y2.decompose_qu2_equ_bqu(tqu,discsize=disc)
```

```python
rteb3=y3.convert_qu2eb(tqu,discsize=disc)
requ3,rbqu3=y3.decompose_qu2_equ_bqu(tqu,discsize=disc)
```

```python
pindex=arange(npix) ; theta=h.pix2ang(nside,pindex)[0]*180/pi
mask=zeros(npix,float)
wband=60
capmask=zeros(npix,float) ; capmask[theta<wband]=1. ; capmask[theta>180-wband]=1.

capmaskon=False ; eqmaskon=True ; nomask=False

if capmaskon:
    eqmaskon=False ; nomaskon=False
    mask[:]=capmask[:]
    figpath= respath + "capmask/" 
    !mkdir $figpath
    h.mollview(mask,title=r"Cap mask $(" + str(wband) + "^\circ)$")
    savefig(figpath + "capmask.pdf",dpi=300,bbox_inches="tight") 
    
if eqmaskon:
    nomask=False
    mask[:]=1- capmask[:]
    figpath=respath + "eqmask/" 
    !mkdir $figpath
    h.mollview(mask,title=r"Equatorial mask $(" + str(wband) + "^\circ)$")
    savefig(figpath + "eqmask.pdf",dpi=300,bbox_inches="tight") 
    
if nomask:
    mask[:]=1. 
    figpath=respath
    h.mollview(mask,title="No mask")
    savefig(figpath + "nomask.pdf",dpi=300,bbox_inches="tight") 

fsky=sum(mask)/size(mask)
print fsky
fll=fll/fsky
```

```python
clteb=h.alm2cl(h.map2alm(teb*mask,lmax=3*nside,pol=False))

clrteb0=h.alm2cl(h.map2alm(rteb0*mask,lmax=3*nside,pol=False))
clrteb1=h.alm2cl(h.map2alm(rteb1*mask,lmax=3*nside,pol=False))
clrteb2=h.alm2cl(h.map2alm(rteb3*mask,lmax=3*nside,pol=False))

clrequ0=h.alm2cl(h.map2alm(requ0*mask,lmax=3*nside,pol=True))
clrequ1=h.alm2cl(h.map2alm(requ1*mask,lmax=3*nside,pol=True))
clrequ2=h.alm2cl(h.map2alm(requ3*mask,lmax=3*nside,pol=True))

clrbqu0=h.alm2cl(h.map2alm(rbqu0*mask,lmax=3*nside,pol=True))
clrbqu1=h.alm2cl(h.map2alm(rbqu1*mask,lmax=3*nside,pol=True))
clrbqu2=h.alm2cl(h.map2alm(rbqu3*mask,lmax=3*nside,pol=True))
```

```python
figure()
plot(ell,fll*clrteb0[1],"r-",lw=1.75,label=r"$\ell_{\rm min}=2$")
plot(ell,fll*clrteb1[1],"b-",lw=1.5,label=r"$\ell_{\rm min}=64$")
plot(ell,fll*clrteb2[1],"g-",lw=1.,label=r"$\ell_{\rm min}=128$")
plot(ell,fll*clteb[1],"k--",lw=1,label="Healpix",alpha=1.)
semilogy()
ylim(1e-2,1e1)
xlim(2,3*nside)
xlabel("multipole,$\ell$")
ylabel("$\ell(\ell+1)C^{EE}_{\ell}/(2\pi) ~ \mu K^2$")
legend(loc=0,fontsize=8,ncol=2,title=r"$r_{\rm cutoff}= 2\beta_o~;~\ell_{\rm max}=192$")
grid(alpha=0.2,linestyle="solid")
vlines(64,1e-3,1e2,linestyle="dashed",alpha=0.6)
vlines(128,1e-3,1e2,linestyle="dashed",alpha=0.6)
if genplots:
    savefig(figpath + "ee-spectrum-rteb-2beta_multipole_filtering.pdf",dpi=300,bbox_inches="tight")
    
figure()
plot(ell,fll*clrteb0[2],"r-",lw=1.75,label=r"$\ell_{\rm min}=2$")
plot(ell,fll*clrteb1[2],"b-",lw=1.5,label=r"$\ell_{\rm min}=64$")
plot(ell,fll*clrteb2[2],"g-",lw=1.,label=r"$\ell_{\rm min}=128$")
plot(ell,fll*clteb[2],"k--",lw=1,label="Healpix",alpha=1.)
semilogy()
ylim(1e-4,1e-1)
xlim(2,3*nside)
xlabel("multipole,$\ell$")
ylabel("$\ell(\ell+1)C^{BB}_{\ell}/(2\pi) ~ \mu K^2$")
legend(loc=0,fontsize=8,ncol=2,title=r"$r_{\rm cutoff}= 2\beta_o~;~\ell_{\rm max}=192$")
grid(alpha=0.2,linestyle="solid")
vlines(64,1e-6,1e2,linestyle="dashed",alpha=0.6)
vlines(128,1e-6,1e2,linestyle="dashed",alpha=0.6)
if genplots:
    savefig(figpath + "bb-spectrum-rteb-2beta_multipole_filtering.pdf",dpi=300,bbox_inches="tight")
```
