```python
def return_euler(theta0,phi0,theta1,phi1):
    cosbeta=sin(theta0)*sin(theta1)*cos(phi1-phi0) + cos(theta0)*cos(theta1)
    beta=arccos(cosbeta)
    
    sa=sin(theta1)*sin(phi1-phi0)#/sqrt(1.-cosbeta*cosbeta)
    ca=(cos(theta0)*sin(theta1)*cos(phi1-phi0)-sin(theta0)*cos(theta1))#/sqrt(1.-cosbeta*cosbeta)
    alpha=arctan2(sa,ca)
    
    sg=-sin(theta0)*sin(phi1-phi0)#/sqrt(1.-cosbeta*cosbeta)
    cg=(cos(theta0)*sin(theta1) - cos(theta1)*sin(theta0)*cos(phi1-phi0))#/sqrt(1.-cosbeta*cosbeta)
    gamma=arctan2(sg,cg)
    return alpha,beta,gamma

def dumb_euler(theta0,phi0,theta1,phi1)
```
