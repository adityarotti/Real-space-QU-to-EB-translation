```python
%pylab notebook
import healpy as h
import numpy.polynomial.legendre as lgdr
import sys  ; sys.path.append("../../codes/")
```

```python
nside=64
ell=arange(3*nside+1)

dtheta=linspace(0.,180.,10000)
cos_dtheta=cos(dtheta*pi/180.)
coeff=((2*ell+1)/(4.*pi))
ctheta=lgdr.legval(cos_dtheta,coeff,tensor=False)
```

```python
plot(dtheta,sin(3*nside*dtheta*pi/180.)/(3*nside*dtheta*pi/180.)**1.5,"--")
plot(dtheta,ctheta/max(ctheta),"r-")
ylim(-0.2,0.2)
xlim(0,20)
```
